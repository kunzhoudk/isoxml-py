```python
import geopandas as gpd
import sys
from pathlib import Path
sys.path.append(str(Path.cwd().parent))

from isoxml.geometry import ShapelyConverterV3
from isoxml.io import read_from_path
from tests.resources.test_resources import TEST_RES_DIR

shp_converter = ShapelyConverterV3()
task_data, _ = read_from_path(Path("/Users/kunzhou/Documents/codebase/isoxml-py/examples/output/big_xml_v3_auto/TASKDATA.XML"))
shp_polys = []
for field in task_data.partfields:
    for poly in field.polygons:
        shp_polys.append(shp_converter.to_shapely_polygon(poly))

field_gs = gpd.GeoSeries(data=shp_polys, crs="EPSG:4326")
field_gs.explore()
```
